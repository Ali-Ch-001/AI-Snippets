```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans


data = pd.read_csv('daily_weather.csv')

features = ['humidity','wind_speed']


x = data[features]

x.dropna(inplace=True)
scaler = StandardScaler()

xs = scaler.fit_transform(x)

inertia_values=[]
k_values = range(1,11)


for i in k_values:
    kmeans = KMeans(n_clusters=i,random_state=42,max_iter=300)
    kmeans.fit(xs)
    inertia_values.append(kmeans.inertia_)


plt.figure(figsize=(8,4))
plt.plot(k_values,inertia_values,marker='o')
plt.title('Elbow')
plt.xlabel('num clus')
plt.ylabel('intertia')
plt.show()


best_k = 4

kmeans = KMeans(n_clusters=best_k,random_state=42,max_iter=300)
kmeans.fit(xs)

data['Cluster'] = kmeans.labels_

cluster_colors = ['red', 'blue', 'green', 'purple']

plt.figure(figsize=(10,7))

for cluster in range(4):
    
    plt.scatter(data[data['Cluster'] == cluster]['humidity'],
                    data[data['Cluster'] == cluster]['wind_speed'],
                    c=cluster_colors[cluster],
                    label=f'Cluster {cluster}' )
    
    
    
plt.title('Kmeans')
plt.xlabel('humidity')
plt.ylabel('wind_speed')
plt.legend()
plt.show()



```
