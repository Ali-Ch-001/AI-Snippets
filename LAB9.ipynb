```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
titanic_df = pd.read_csv('train.csv')

# Passenger Demographics
# 1. Distribution of passengers across different passenger classes (Pclass)
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=titanic_df)
plt.title('Distribution of Passengers by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()

# 2. Age distribution of passengers
plt.figure(figsize=(8, 6))
sns.histplot(titanic_df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# 3. Gender distribution among passengers
plt.figure(figsize=(6, 6))
sns.countplot(x='Sex', data=titanic_df)
plt.title('Gender Distribution Among Passengers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Survival Analysis
# 1. Overall survival rate of passengers
plt.figure(figsize=(6, 6))
sns.countplot(x='Survived', data=titanic_df)
plt.title('Overall Survival Rate of Passengers')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.show()

# 2. Relationship between passenger class and survival rate
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', data=titanic_df)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# 3. Correlation between age and survival
plt.figure(figsize=(8, 6))
sns.barplot(x='Age', y='Survived', data=titanic_df)
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.show()

# 4. Difference in survival rates based on gender
plt.figure(figsize=(6, 6))
sns.barplot(x='Sex', y='Survived', data=titanic_df)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()

# 5. Effect of having family members aboard on survival chances
titanic_df['Family'] = titanic_df['SibSp'] + titanic_df['Parch']
plt.figure(figsize=(8, 6))
sns.barplot(x='Family', y='Survived', data=titanic_df)
plt.title('Survival Rate by Family Size')
plt.xlabel('Family Size')
plt.ylabel('Survival Rate')
plt.show()

# Ticket Fare and Cabin
# 1. Distribution of ticket fares
plt.figure(figsize=(8, 6))
sns.histplot(titanic_df['Fare'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Ticket Fares')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()

# 2. Correlation between ticket fare and passenger class
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Fare', data=titanic_df)
plt.title('Average Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.show()

# 3. Relationship between availability of cabin data and survival rates
titanic_df['HasCabin'] = titanic_df['Cabin'].notnull()
plt.figure(figsize=(6, 6))
sns.barplot(x='HasCabin', y='Survived', data=titanic_df)
plt.title('Survival Rate based on Cabin Availability')
plt.xlabel('Has Cabin')
plt.ylabel('Survival Rate')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.show()

# Embarkation Port
# 1. Distribution of passengers based on the port of embarkation
plt.figure(figsize=(6, 6))
sns.countplot(x='Embarked', data=titanic_df)
plt.title('Distribution of Passengers by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Count')
plt.show()

# 2. Correlation between the embarkation port and survival rates
plt.figure(figsize=(6, 6))
sns.barplot(x='Embarked', y='Survived', data=titanic_df)
plt.title('Survival Rate by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Survival Rate')
plt.show()
```
